```python
#1: Dibuja la gráfica usando una línea roja gruesa.

x = var("x")
f(x) = sin(pi*x - pi)
plot(f(x), -1, 1, color = 'red', thickness = 5)
```

```python
#2: Dibuja la gráfica usando una línea gruesa azul.

n = var("n")
f(n) = cos(pi*n - pi)
plot(f(n), -1, 1, color = 'blue', thickness = 5)
```

```python
#3: Dibuja las dos curvas anteriores sobre el mismo sistema de ejes.

plot(sin(pi*x - pi), -1, 1, color = 'red', thickness = 5) + plot(cos(pi*n - pi), -1, 1, color = 'blue', thickness = 5)

        # Donde la prumera función es de color rojo y la segunda de color azul.
        ## Si presento las ecuaciones antes y después pongo el f(x) / f(n) dentro de la fórmula solo se proyecta le segunda.
```

```python
#4: Dibuja la gráfica de "y" ajustando el rango.

x = var("x")
f(x) = 1 / x
plot(f(x), -1, 1, ymin=-10, ymax=10, color = 'black', thickness = 5)
```

```python
#5: Dibuja y · sin(x^2 - y^2) = x · cos (x + y), el rango determinado.

x, y = var("x, y")
f(x, y) = ([y * sin(x^2 - y^2) == x * cos(x + y)], x, y)
plot(f(x, y), (x,-3, 3), ymin=-3, ymax=3, thickness = 1)
```
